```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
courses = ["course 1", "course 2","course 3","course 4","course 5","course 6"]
marks = [60, 40, 80, 90, 100, 50]
```

```python
plt.bar(courses,marks, width=0.5)
plt.show()
```

```python
plt.plot(courses, marks)
plt.show()
```

```python
plt.scatter(courses[0], marks[0], label="SET-1")
plt.scatter(courses[1], marks[1], label="SET-2")
plt.scatter(courses[2], marks[2], label="SET-3")
plt.scatter(courses[3], marks[3], label="SET-4")
plt.scatter(courses[4], marks[4], label="SET-5")
plt.title("MY PLOT")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend();
plt.show()
```

```python
x1 = np.arange(100)
y1 =  np.random.randint(1,70,100)
x2 = np.arange(80)
y2 =  np.random.randint(1,100,80)
x3 = np.arange(80)
y3 =  np.random.randint(1,60,80)
plt.scatter(x1, y1, c="blue")
plt.scatter(x2, y2, c ="red")
plt.scatter(x3, y3, c ="green")
plt.title("MY PLOT")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
